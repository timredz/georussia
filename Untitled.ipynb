### populated places

```python
f = open('populated_places.json' ,encoding='utf8')
 
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
features = []
for i in range(len(data['features'])):
    if (data['features'][i]['properties']['SOV0NAME'] == 'Russia') and (data['features'][i]['properties']['SCALERANK'] <= 6):
        features.append(data['features'][i])
        
data['features'] = features
```

```python
for i in range(len(data['features'])):
    p = data['features'][i]['properties']
    new_p = {'SOV0NAME': p['SOV0NAME'], 'ADM0_A3': p['ADM0_A3'], 'NAME_RU': p['NAME_RU'], 'RANK': p['SCALERANK']}
    data['features'][i]['properties'] = new_p
```

```python
cities = data
len(data['features'])
```

```python
all_obj = {'type': 'FeatureCollection', 'features': countries['features'] + cities['features']}
```

```python
json_object = json.dumps(all_obj, indent=4)
```

```python
with open("russia.geo.json", "w") as outfile:
    outfile.write(json_object)
```
